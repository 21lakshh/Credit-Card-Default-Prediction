```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
credit_data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/IIT HACKATHON/Dev_data_to_be_shared.csv")
```

```python
credit_data.head()
```

```python
# On-us attributes: Related to the customer’s credit limit or the credit utilization
# Transaction attributes: Represent customer transaction patterns (e.g., number of transactions, transaction value across merchant types).
# Bureau tradeline attributes: Historical delinquency and product holding details.
# Bureau inquiry attributes: Represent recent credit inquiries reported by credit bureaus (e.g., number of personal loan inquiries in the past 3 months).
```

```python
credit_data.info()
```

```python
credit_data.describe()
```

```python
credit_data.isnull().sum()
```

```python
bureau_enquiry_columns = []
for col in credit_data.columns:
    if col.startswith("bureau_enquiry"):
        bureau_enquiry_columns.append(col)

for col in credit_data.columns:
    if col not in bureau_enquiry_columns:
      credit_data[col].fillna(credit_data[col].mean(), inplace=True)

for col in bureau_enquiry_columns:
    credit_data[col].fillna(0, inplace=True)
```

```python
credit_data.isnull().sum()
```

```python
null_counts = credit_data.isnull().sum()
print("Null values in each column:")
print(null_counts[null_counts > 0])
```

```python
# these columns are fully NULL so we replace them with 0
credit_data['bureau_436'].fillna(0, inplace=True)
credit_data['bureau_447'].fillna(0, inplace=True)
```

```python
null_counts = credit_data.isnull().sum()
print("Null values in each column:")
print(null_counts[null_counts > 0])
```

```python
credit_data.head()
```

```python
credit_data['bad_flag'].value_counts()
```

```python
X = credit_data.drop(columns=["bad_flag", "account_number"])  # Drop target and primary key
y = credit_data["bad_flag"]
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled
```

```python
# Splitting data into training and validation
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from xgboost import XGBRegressor
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
models = [LogisticRegression(), DecisionTreeClassifier(), RandomForestClassifier(), GradientBoostingClassifier(), KNeighborsClassifier(), XGBClassifier()]
```

```python
import sklearn.metrics as mt
from sklearn import metrics
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
```

```python
best_score = 0
best_model = None
for model in models:
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    score = model.score(X_test, y_test)
    print(f"{model.__class__.__name__}:")
    print("Accuracy:", mt.accuracy_score(y_test, y_pred))
    print("Precision:", mt.precision_score(y_test, y_pred, average='weighted'))
    print("Recall:", mt.recall_score(y_test, y_pred, average='weighted'))
    print("F1-score:", mt.f1_score(y_test, y_pred, average='weighted'))
    print("Score:", score)
    print()
    if score > best_score:
        best_score = score
        best_model = model
    print("Best model updated:", best_model)
```

```python
print("Best Model:", best_model.__class__.__name__, ", Best Score:", best_score)
```
